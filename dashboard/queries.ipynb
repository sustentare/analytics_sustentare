## Table Join

```python
def query_join(connection):

    if connection is None:
        return "Connection to database failed."

    cursor = connection.cursor()
    cursor.execute("SHOW TABLES")
    tables = cursor.fetchall()

    query = f''' SELECT p.cd_pessoa, m.codigoaluno, p.nm_pessoa, m.anosemestre, m.turma, m.curso,
        m.dataemissao, m.usuario, m.planopagamento, m.cd_plano_pgto
        
    FROM unimestre.pessoas AS p
    INNER JOIN unimestre.matriculas AS m ON p.cd_pessoa = m.codigoaluno
    WHERE YEAR(m.dataemissao) >= 2000 AND NOT m.turma REGEXP '^[0-9]+$'
    ORDER BY m.dataemissao DESC;

'''
    df = pd.read_sql(query, connection)

    cursor.close()

    return df

df_query_join = query_join(db_connection)
```

# Join Mensalidades & Turmas

```python
def unimestre_join(connection):

    if connection is None:
        return "Connection to database failed."

    cursor = connection.cursor()
    cursor.execute("SHOW TABLES")
    tables = cursor.fetchall()

    query = f'''SELECT mensal.codigoaluno, mensal.valorbruto, mensal.descontoextra, 
    mensal.VALORTOTAL, mensal.valorpago, mensal.curso as mensalidade_curso, 
    mensal.turma as mensalidade_turma, p.nm_pessoa
FROM unimestre.mensalidades AS mensal
INNER JOIN unimestre.pessoas AS p ON mensal.codigoaluno = p.cd_pessoa
GROUP BY
    mensal.codigoaluno, mensal.turma
LIMIT 10;
'''
    df = pd.read_sql(query, connection)

    cursor.close()

    return df

df_unimestre_join = unimestre_join(db_connection)
```
